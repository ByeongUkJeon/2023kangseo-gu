```python
tmp_data = pop[['65세이상고령자','기초생활수급권자']]

scaler = MinMaxScaler()

data_scale = scaler.fit_transform(tmp_data)
```

```python
import matplotlib.pyplot as plt
```

```python
def visualize_elbowmethod(data, param_init='random', param_n_init=10, param_max_iter=300):
    distortions = []
    for i in range(1, 10):
        km = KMeans(n_clusters=i, init=param_init, n_init=param_n_init, max_iter=param_max_iter, random_state=0)
        km.fit(data)
        distortions.append(km.inertia_)
        
    plt.rc('font', family='Malgun Gothic')
    plt.plot(range(1, 10), distortions, marker='o')
    plt.xlabel('Number of Cluster')
    plt.ylabel('Distortion')
    plt.show()
    
visualize_elbowmethod(tmp_data)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import seaborn as sns
def visualize_silhouette_layer(data, param_init='random', param_n_init=10, param_max_iter=300):
    clusters_range = range(2,10)
    results = []

    for i in clusters_range:
        clusterer = KMeans(n_clusters=i, init=param_init, n_init=param_n_init, max_iter=param_max_iter, random_state=0)
        cluster_labels = clusterer.fit_predict(data)
        silhouette_avg = silhouette_score(data, cluster_labels)
        results.append([i, silhouette_avg])

    result = pd.DataFrame(results, columns=["n_clusters", "silhouette_score"])
    pivot_km = pd.pivot_table(result, index="n_clusters", values="silhouette_score")

    plt.figure()
    sns.heatmap(pivot_km, annot=True, linewidths=.5, fmt='.3f', cmap=sns.cm._rocket_lut)
    plt.tight_layout()
    plt.show()

visualize_silhouette_layer(tmp_data)
```

```python
from sklearn.cluster import KMeans
from tensorflow.keras import backend

backend.clear_session()

k = 7

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 10)

# 정규화된 데이터에 학습
model.fit(data_scale)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
pop['cluster'] = model.fit_predict(data_scale)
```

```python
pop.to_csv('gangseo_hangjeong.csv')
```

```python
plt.figure(figsize = (8, 8))
plt.rc('font', family='Malgun Gothic')
for i in range(k):
    plt.scatter(pop.loc[pop['cluster'] == i, '기초생활수급권자'], pop.loc[pop['cluster'] == i, '65세이상고령자'], 
                label = 'cluster ' + str(i))

plt.legend()
plt.title('K = %d results'%k , size = 15)
plt.xlabel('기초생활수급권자', size = 12)
plt.ylabel('65세이상고령자', size = 12)
plt.show()
```
