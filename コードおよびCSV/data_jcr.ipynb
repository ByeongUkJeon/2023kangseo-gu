```python
from sklearn.preprocessing import MinMaxScaler

tmp_data = gangseo_info[['인구대비공원수', '평균집값(만원)', '기초생활수급권자']]

scaler = MinMaxScaler()

data_scale = scaler.fit_transform(tmp_data)
```

```python
from sklearn.cluster import KMeans

k = 4

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 10)

# 정규화된 데이터에 학습
model.fit(data_scale)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
gangseo_info['cluster'] = model.fit_predict(data_scale)
```

```python
plt.figure(figsize = (8, 8))

for i in range(k):
    plt.scatter(gangseo_info.loc[gangseo_info['cluster'] == i, '인구대비공원수'], gangseo_info.loc[gangseo_info['cluster'] == i, '기초생활수급권자'], 
                label = 'cluster ' + str(i))

plt.legend()
plt.title('K = %d results'%k , size = 15)
plt.xlabel('인구대비공원수', size = 12)
plt.ylabel('기초생활수급권자', size = 12)
plt.show()
```
